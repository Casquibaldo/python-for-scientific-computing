```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

ax = plt.figure().add_subplot(111)
data = np.random.normal(5.0,3.0,1000)
ax.hist(data,bins=30)
ax.set_xlabel('data')
ax.set_ylabel('numer of values')
plt.show()
```

```python
# same thing without numpy:

def getMin(data):
    minimum = data[0]
    for i in range(len(data)):
        if minimum > data[i]:
            minimum = data[i]
    return minimum

def getMax(data):
    maximum = data[0]
    for i in range(len(data)):
        if maximum < data[i]:
            maximum = data[i]
    return maximum

def getHistogram(data, numBins):
    maximum = getMax(data)
    minimum = getMin(data)
    # the integer part of following division will always be less than numBins
    deltaBin = (maximum*1.0001 - minimum)/numBins
    bincounts = [0 for i in range(numBins)]
    for i in range(len(data)):
        bincounts[int((data[i] - minimum)/deltaBin)] += 1
    return minimum, maximum, bincounts

ax = plt.figure().add_subplot(111)

numsamples = 1000
numBins = 30

# we're gonna use numpy for the numbers though...
d = np.random.normal(5.0,3.0,numsamples)

# I'm feeding in a list, just to make sure my code works for lists
m, M, h = getHistogram(list(d),numBins)
db = (M-m) / numBins
b = [db*i + m for i in range(numBins)]
ax.plot(b, h)
# and check against numpy version:
ax.hist(d, bins = numBins, histtype = 'step')

ax.set_xlabel('data')
ax.set_ylabel('numer of values')
plt.show()
```
