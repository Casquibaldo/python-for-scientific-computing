```python
x = np.array([2,3], dtype = np.float)
y = np.array([5,5], dtype = np.float)

def projection(x,y):
    scale = np.sum(x*y, axis = 0)/np.sum(y**2,axis = 0)
    scale2 = np.dot(x,y)/np.dot(y,y)
    proj = scale2*y
    return proj

z = projection(x,y)
print z
```

```python
ax = plt.figure().add_subplot(111)
ax.arrow(
    0, 0, x[0],x[1],
    color = 'red',
    width = 0.01,
    length_includes_head =True)
ax.arrow(
    0, 0, y[0],y[1],
    color = 'black',
    width = 0.01,
    length_includes_head =True)
ax.arrow(
    0, 0, z[0],z[1],
    color = 'blue',
    width = 0.01,
    length_includes_head =True)
ax.set_xlim(0,6)
ax.set_ylim(0,6)
ax.set_aspect('equal')
```

```python
t = np.linspace(0, 2*np.pi, 2**10)

x = .9*np.cos(3*t)
y = .9*np.sin(4*t)
z = t**2
```

```python
fig = plt.figure()


ax = fig.add_subplot(221)
a = np.array([1,0,0], dtype = np.float)
b = np.array([0,1,0], dtype = np.float)
X = a[0]*x + a[1]*y + a[2]*z
Y = b[0]*x + b[1]*y + b[2]*z
X /= (np.sum(a**2))**(.5)
Y /= (np.sum(b**2))**(.5)
ax.plot(X,Y)



ax = fig.add_subplot(222)
a = np.array([0,0,1], dtype = np.float)
b = np.array([0,1,0], dtype = np.float)
X = a[0]*x + a[1]*y + a[2]*z
Y = b[0]*x + b[1]*y + b[2]*z
X /= (np.sum(a**2))**(.5)
Y /= (np.sum(b**2))**(.5)
ax.plot(X,Y)

ax = fig.add_subplot(223)
a = np.array([1,0,.5], dtype = np.float)
b = np.array([-.5,1,0], dtype = np.float)
X = a[0]*x + a[1]*y + a[2]*z
Y = b[0]*x + b[1]*y + b[2]*z
X /= (np.sum(a**2))**(.5)
Y /= (np.sum(b**2))**(.5)
ax.plot(X,Y)

ax = fig.add_subplot(224)
a = np.random.random(3)
b = np.random.random(3)
X = a[0]*x + a[1]*y + a[2]*z
Y = b[0]*x + b[1]*y + b[2]*z
X /= (np.sum(a**2))**(.5)
Y /= (np.sum(b**2))**(.5)
ax.plot(X,Y)

```
