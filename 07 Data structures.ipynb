\begin{equation}
    f_c: \mathbb{C} \rightarrow {\mathbb{C}}, f_c(z) = z^2 + c
\end{equation}


we are interested in the bounded set $M$ such that $f_c^n(0) \in M$ ($f_c$ composed with itself $n$ times) for all $n \in \mathbb{N}$, for all $c \in M$.

```python
import numpy as np

x = np.linspace(-2, 2, 256)
y = np.linspace(2, 0, 128)

z = x[np.newaxis, :] + 1j*y[:, np.newaxis]
iterations = np.zeros(z.shape, np.int)

def quadratic_map(z, c = 0):
    return z**2 + c

for j in range(y.shape[0]):
    for i in range(x.shape[0]):
        z0 = 0
        nmax = 0
        for n in range(1, 20):
            if abs(z0) > 2:
                break
            z0 = quadratic_map(z0, z[j, i])
            nmax = n
        iterations[j, i] = nmax
```

```python
%matplotlib nbagg
import matplotlib.pyplot as plt

ax = plt.figure().add_subplot(211)
ax.imshow(iterations, interpolation = 'none', extent = [x[0], x[-1], y[-1], y[0]])

iter2 = np.zeros(iterations.shape, iterations.dtype)
indices = []
for i in range(10):
    indices.append(np.where(iterations == i))
iter2[indices[6]] = 1

ax = plt.gcf().add_subplot(212)
ax.imshow(iter2, interpolation = 'none', extent = [x[0], x[-1], y[-1], y[0]])
```

```python
import numpy as np

x = np.linspace(-2, 2, 1024)
y = np.linspace(2, 0, 512)

c = x[np.newaxis, :] + 1j*y[:, np.newaxis]
z = np.zeros(c.shape, c.dtype)
iterations = np.zeros(z.shape, np.int)

def quadratic_map(z, c = 0):
    return z**2 + c

for n in range(1, 40):
    good_indices = np.where(np.abs(z) < 2)
    z[good_indices] = z[good_indices]**2 + c[good_indices]
    iterations[good_indices] += 1

ax = plt.figure(figsize=(13, 13)).add_subplot(211)
ax.imshow(iterations, interpolation = 'none', extent = [x[0], x[-1], y[-1], y[0]])
```
