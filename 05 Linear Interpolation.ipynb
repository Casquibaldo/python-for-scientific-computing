linear interpolation:
---------------------

we are given $x_0$, $x_1$, and the values that some function takes at these points, $y_0$ and $y_1$ respectively.
for an arbitrary $x_0 \leq x \leq x_1$, the linear approximation to that function is

\begin{equation}
    y = y_0 + \frac{y_1 - y_0}{x_1 - x_0}(x - x_0)
\end{equation}

```python
def interp(tt):
    for i in range(t.shape[0]):
        if tt < t[i]:
            i0 = i
            break
        else:
            i0 = i
    y = x[i0] + (x[i0] - x[i0-1])/(t[i0] - t[i0-1]) * (tt - t[i0])
    return y
import numpy as np
tfine = np.linspace(0, 1, 50)
yfine = []
for var in tfine:
    yfine.append(interp(var))

ax = plt.figure(figsize=(6, 3)).add_subplot(111)
ax.plot(t, x, label = 'coarse data')
ax.plot(tfine, yfine, marker = '+', linestyle = 'none', label = 'interpolated data')
ax.legend(loc = 'best')
```

```python
class interpolator:
    # members:
    #  t -- array of function arguments where we know the value of the function
    #  x -- array of corresponding values of the function
    # methods:
    #  interpolate -- linear interpolation
    def __init__(self, t, x):
        self.t = t
        self.x = x
        return None
    def interpolate(self, tt):
        for i in range(self.t.shape[0]):
            if tt < self.t[i]:
                i0 = i
                break
            else:
                i0 = i
        y = (self.x[i0] +
             ((self.x[i0] - self.x[i0-1]) /
              (self.t[i0] - self.t[i0-1])) * (tt - self.t[i0]))
        return y
    

farguments = np.linspace(0, 1, 17)
fvalues    = np.sin(2*np.pi*farguments) + np.cos(3*np.pi*farguments)

i = interpolator(farguments, fvalues)

tfine = np.linspace(0, 1, 50)
yfine = []
for var in tfine:
    yfine.append(i.interpolate(var))

ax = plt.figure(figsize=(6, 3)).add_subplot(111)
ax.plot(farguments, fvalues, marker = '.', label = 'coarse data')
ax.plot(tfine, yfine, marker = '+', linestyle = 'none', label = 'interpolated data')
ax.legend(loc = 'best')
```

Multivariate interpolation
--------------------------

```python
class interpolator2D:
    # members:
    #  x -- array of function arguments 0 where we know the value of the function
    #  y -- array of function arguments 1 where we know the value of the function
    #  f -- array of corresponding values of the function
    # methods:
    #  interpolate -- linear interpolation
    def __init__(self, x, y, f):
        assert(x.shape[0] == f.shape[1])
        assert(y.shape[0] == f.shape[0])
        self.x = x
        self.y = y
        self.f = f
        return None
    def find_index(self, xx, array):
        if (xx < array[0]):
            i0 = 0
        for i in range(array.shape[0]-1):
            if xx >= array[i] and xx < array[i+1]:
                i0 = i
        if (xx > array[array.shape[0]-1]):
            i0 = array.shape[0]-1
        return i0
    def interpolate(self, xx, yy):
        i0 = self.find_index(xx, self.x)
        j0 = self.find_index(yy, self.y)
        fy0 = (self.f[j0, i0] +
             ((self.f[j0, i0+1] - self.f[j0, i0]) /
              (self.x[i0+1] - self.x[i0])) * (xx - self.x[i0]))
        fy1 = (self.f[j0+1, i0] +
             ((self.f[j0+1, i0+1] - self.f[j0+1, i0]) /
              (self.x[i0+1] - self.x[i0])) * (xx - self.x[i0]))
        return (fy0 +
             ((fy1 - fy0) /
              (self.y[j0+1] - self.y[j0])) * (yy - self.y[j0]))
```

```python

x = np.linspace(0, 3., 192)
y = np.linspace(0, 2., 128)
# broadcasting a numpy array: making an extra dimension, and making copies of array along that dimension
z = (np.sin(  np.pi*x[np.newaxis, :])*np.cos(2*np.pi*y[:, np.newaxis]) +
     np.cos(3*np.pi*x[np.newaxis, :])*np.sin( np.pi*y[:, np.newaxis]))
print x.shape
print(y.shape)
print(x[np.newaxis, :].shape)
print(y[:, np.newaxis].shape)
print(z.shape)

ax = plt.figure().add_subplot(111)
ax.imshow(z, interpolation = 'none') # interpolation can be none, bilinear, or bicubic
```

```python
def func(xx, yy):
    return (np.sin(  np.pi*xx)*np.cos(2*np.pi*yy) +
            np.cos(3*np.pi*xx)*np.sin(  np.pi*yy))

interp = interpolator2D(x, y, z)
err = []
for i in range(100):
    xx = x[-1]*np.random.random()
    yy = y[-1]*np.random.random()
    err.append(abs(func(xx, yy) - interp.interpolate(xx, yy)))

print np.average(err)/np.average(np.abs(z))
```

If something is conceptually a function, then go ahead and use it like a function.
Don't be afraid to overload operators, including the "function call" operator "()".
Or `__call__` in python speak.

```python
class interpolator2D:
    # members:
    #  x -- array of function arguments 0 where we know the value of the function
    #  y -- array of function arguments 1 where we know the value of the function
    #  f -- array of corresponding values of the function
    # methods:
    #  interpolate -- linear interpolation
    def __init__(self, x, y, f):
        assert(x.shape[0] == f.shape[1])
        assert(y.shape[0] == f.shape[0])
        self.x = x
        self.y = y
        self.f = f
        return None
    def find_index(self, xx, array):
        if (xx < array[0]):
            i0 = 0
        for i in range(array.shape[0]-1):
            if xx >= array[i] and xx < array[i+1]:
                i0 = i
        if (xx > array[array.shape[0]-1]):
            i0 = array.shape[0]-1
        return i0
    def __call__(self, xx, yy):
        i0 = self.find_index(xx, self.x)
        j0 = self.find_index(yy, self.y)
        fy0 = (self.f[j0, i0] +
             ((self.f[j0, i0+1] - self.f[j0, i0]) /
              (self.x[i0+1] - self.x[i0])) * (xx - self.x[i0]))
        fy1 = (self.f[j0+1, i0] +
             ((self.f[j0+1, i0+1] - self.f[j0+1, i0]) /
              (self.x[i0+1] - self.x[i0])) * (xx - self.x[i0]))
        return (fy0 +
             ((fy1 - fy0) /
              (self.y[j0+1] - self.y[j0])) * (yy - self.y[j0]))
    
def func(xx, yy):
    return (np.sin(  np.pi*xx)*np.cos(2*np.pi*yy) +
            np.cos(3*np.pi*xx)*np.sin(  np.pi*yy))

func_linear = interpolator2D(x, y, z)
err = []
for i in range(100):
    xx = x[-1]*np.random.random()
    yy = y[-1]*np.random.random()
    err.append(abs(func(xx, yy) - func_linear(xx, yy)))

print np.average(err)/np.average(np.abs(z))
```
