Lagrange interpolation
----------------------

```python
import numpy as np
%matplotlib nbagg
from matplotlib import pyplot as plt


t = np.linspace(0, 1, 10)
x = np.sin(2*np.pi*t)
ax = plt.figure(figsize=(6, 3)).add_subplot(111)
ax.plot(t, x, marker = '.')
```

we are given the real numbers $x_0$, $x_1$, ..., $x_{n-1}$,
and the real numbers $y_0$, $y_1$, ..., $y_{n-1}$.

there's an infinity of polynomials $P(x)$ such that $P(x_i) = y_i$.


\begin{equation}
    P(x) = \sum_{i=0}^m a_i x^i, m \geq n   
\end{equation}

\begin{equation}
    \sum_{p=0}^m a_p x_i^p = y_i, 0 \leq i < n
\end{equation}

```python
class interpolator2D:
    # members:
    #  x -- array of function arguments 0 where we know the value of the function
    #  y -- array of function arguments 1 where we know the value of the function
    #  f -- array of corresponding values of the function
    # methods:
    #  interpolate -- linear interpolation
    def __init__(self, x, y, f):
        assert(x.shape[0] == f.shape[1])
        assert(y.shape[0] == f.shape[0])
        self.x = x
        self.y = y
        self.f = f
        return None
    def find_index(self, xx, array):
        if (xx < array[0]):
            i0 = 0
        for i in range(array.shape[0]-1):
            if xx >= array[i] and xx < array[i+1]:
                i0 = i
        if (xx > array[array.shape[0]-1]):
            i0 = array.shape[0]-1
        return i0
    def xlin(self, p0, node, value):
        if node == 0:
            xval = 1 - (value - self.x[p0]) / (self.x[p0+1] - self.x[p0])
        if node == 1:
            xval = (value - self.x[p0]) / (self.x[p0+1] - self.x[p0])
        return xval
    def ylin(self, p0, node, value):
        if node == 0:
            xval = 1 - (value - self.y[p0]) / (self.y[p0+1] - self.y[p0])
        if node == 1:
            xval = (value - self.y[p0]) / (self.y[p0+1] - self.y[p0])
        return xval
    def __call__(self, xx, yy):
        i0 = self.find_index(xx, self.x)
        j0 = self.find_index(yy, self.y)
        return ((self.f[j0  , i0  ]*self.xlin(i0, 0, xx) +
                 self.f[j0  , i0+1]*self.xlin(i0, 1, xx))*self.ylin(j0, 0, yy) +
                (self.f[j0+1, i0  ]*self.xlin(i0, 0, xx) +
                 self.f[j0+1, i0+1]*self.xlin(i0, 1, xx))*self.ylin(j0, 1, yy))
    
def func(xx, yy):
    return (np.sin(  np.pi*xx)*np.cos(2*np.pi*yy) +
            np.cos(3*np.pi*xx)*np.sin(  np.pi*yy))


x = np.linspace(0, 3., 192)
y = np.linspace(0, 2., 128)

z = func(x[np.newaxis, :], y[:, np.newaxis])

func_linear = interpolator2D(x, y, z)
err = []
for i in range(100):
    xx = x[-1]*np.random.random()
    yy = y[-1]*np.random.random()
    err.append(abs(func(xx, yy) - func_linear(xx, yy)))

print np.average(err)/np.average(np.abs(z))
```

```python
class interpolator2D:
    # members:
    #  x -- array of function arguments 0 where we know the value of the function
    #  y -- array of function arguments 1 where we know the value of the function
    #  f -- array of corresponding values of the function
    # methods:
    #  interpolate -- linear interpolation
    def __init__(self, x, y, f):
        assert(x.shape[0] == f.shape[1])
        assert(y.shape[0] == f.shape[0])
        self.x = x
        self.y = y
        self.f = f
        return None
    def find_index(self, xx, array):
        if (xx < array[0]):
            i0 = 0
        for i in range(array.shape[0]-1):
            if xx >= array[i] and xx < array[i+1]:
                i0 = i
        if (xx > array[array.shape[0]-1]):
            i0 = array.shape[0]-1
        return i0
    def xlin(self, p0, node, value):
        if node == 0:
            xval = -(value - self.x[p0+1]) / (self.x[p0+1] - self.x[p0])
        if node == 1:
            xval = (value - self.x[p0  ]) / (self.x[p0+1] - self.x[p0])
        return xval
    def ylin(self, p0, node, value):
        if node == 0:
            xval = -(value - self.y[p0+1]) / (self.y[p0+1] - self.y[p0])
        if node == 1:
            xval = (value - self.y[p0]) / (self.y[p0+1] - self.y[p0])
        return xval
    def polyLag3(self, array, p0, node, value):
        xval = 1.0
        for m in range(p0-1, p0+3):
            if not (m == p0 + node):
                xval *= (value - array[m]) / (array[p0+node] - array[m])
        return xval
    def __call__(self, xx, yy):
        i0 = self.find_index(xx, self.x)
        j0 = self.find_index(yy, self.y)
        return ((self.f[j0  , i0  ]*self.xlin(i0, 0, xx) +
                 self.f[j0  , i0+1]*self.xlin(i0, 1, xx))*self.ylin(j0, 0, yy) +
                (self.f[j0+1, i0  ]*self.xlin(i0, 0, xx) +
                 self.f[j0+1, i0+1]*self.xlin(i0, 1, xx))*self.ylin(j0, 1, yy))
    def Lag3(self, xx, yy):
        i0 = self.find_index(xx, self.x)
        j0 = self.find_index(yy, self.y)
        value = 0.0
        for j in range(-1, 3):
            for i in range(-1, 3):
                value += (self.f[j0 + j, i0 + i] *
                          self.polyLag3(self.x, i0, i, xx) *
                          self.polyLag3(self.y, j0, j, yy))
        return value


def func(xx, yy):
    return (np.sin(  np.pi*xx)*np.cos(2*np.pi*yy) +
            np.cos(3*np.pi*xx)*np.sin(  np.pi*yy))


x = np.linspace(0, 3., 72)
y = np.linspace(0, 2., 48)

z = func(x[np.newaxis, :], y[:, np.newaxis])

func_linear = interpolator2D(x, y, z)
err1 = []
err3 = []
for i in range(100):
    xx = x[2] + (x[-3] - x[2])*np.random.random()
    yy = y[2] + (y[-3] - y[2])*np.random.random()
    #yy = y[-1]*np.random.random()
    err1.append(abs(func(xx, yy) - func_linear(xx, yy)))
    err3.append(abs(func(xx, yy) - func_linear.Lag3(xx, yy)))

print np.average(err1)/np.average(np.abs(z))
print np.average(err3)/np.average(np.abs(z))
```
