linear interpolation:
---------------------

we are given $x_0$, $x_1$, and the values that some function takes at these points, $y_0$ and $y_1$ respectively.
for an arbitrary $x_0 \leq x \leq x_1$, the linear approximation to that function is

\begin{equation}
    y = y_0 + \frac{y_1 - y_0}{x_1 - x_0}(x - x_0)
\end{equation}

```python
def interp(tt):
    for i in range(t.shape[0]):
        if tt < t[i]:
            i0 = i
            break
        else:
            i0 = i
    y = x[i0] + (x[i0] - x[i0-1])/(t[i0] - t[i0-1]) * (tt - t[i0])
    return y
import numpy as np
tfine = np.linspace(0, 1, 50)
yfine = []
for var in tfine:
    yfine.append(interp(var))

ax = plt.figure(figsize=(6, 3)).add_subplot(111)
ax.plot(t, x, label = 'coarse data')
ax.plot(tfine, yfine, marker = '+', linestyle = 'none', label = 'interpolated data')
ax.legend(loc = 'best')
```

```python
class interpolator:
    # members:
    #  t -- array of function arguments where we know the value of the function
    #  x -- array of corresponding values of the function
    # methods:
    #  interpolate -- linear interpolation
    def __init__(self, t, x):
        self.t = t
        self.x = x
        return None
    def interpolate(self, tt):
        for i in range(self.t.shape[0]):
            if tt < self.t[i]:
                i0 = i
                break
            else:
                i0 = i
        y = (self.x[i0] +
             ((self.x[i0] - self.x[i0-1]) /
              (self.t[i0] - self.t[i0-1])) * (tt - self.t[i0]))
        return y
    

farguments = np.linspace(0, 1, 17)
fvalues    = np.sin(2*np.pi*farguments) + np.cos(3*np.pi*farguments)

i = interpolator(farguments, fvalues)

tfine = np.linspace(0, 1, 50)
yfine = []
for var in tfine:
    yfine.append(i.interpolate(var))

ax = plt.figure(figsize=(6, 3)).add_subplot(111)
ax.plot(farguments, fvalues, marker = '.', label = 'coarse data')
ax.plot(tfine, yfine, marker = '+', linestyle = 'none', label = 'interpolated data')
ax.legend(loc = 'best')
```
